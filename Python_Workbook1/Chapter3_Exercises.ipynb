```python
# Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

# Enter Hours: 45
# Enter Rate: 10
# Pay: 475.0


hours = int(input("Enter hours: "))
rate = float(input("Enter rate: "))

if hours > 40:
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * 1.5 * rate
    regular_pay = 40 * rate
    total_pay = regular_pay + overtime_pay
    print("Pay:", total_pay)
else:
    pay = hours * rate
    print("Pay:", pay)


```

```python
# Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

# Score           Grade
# [ 97, 100]      1.0
# [ 93,  97)        1.25
# [ 89,  93)        1.5
# [ 85,  89)        1.75
# [ 81,  85)        2.0
# [ 77,  81)        2.25
# [ 74,  77)        2.5
# [ 0,   74)        3.0


score = int(input("Enter score between 0 and 100: "))

if(score >= 97):
    print("Grade: 1.0")
elif(score >= 93):
    print("Grade: 1.25")
elif(score >= 89):
    print("Grade: 1.5")
elif(score >= 85):
    print("Grade: 1.75")
elif(score >= 81):
    print("Grade: 2.0")
elif(score >= 77):
    print("Grade:2.25")
elif(score >= 74):
    print("Grade: 2.5")
elif(score >= 0):
    print("Grade: 3.0")
else :
    print("Grade: None")
```
