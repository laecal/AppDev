```python
# Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

# Enter Hours: 45
# Enter Rate: 10
# Pay: 475.0

def compute_pay(hours, rate):
  if hours > 40:
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * 1.5 * rate
    regular_pay = 40 * rate
    total_pay = regular_pay + overtime_pay
    return total_pay
  else:
    pay = hours * rate
    return pay

hours = int(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

payment = compute_pay(hours, rate)
print("Pay: ", payment)

```

```python
# Write a function called computegrade that takes a score as its
# parameter and returns a grade as a string using the following table:

# Score           Grade
# [ 97, 100]      1.0
# [ 93,  97)        1.25
# [ 89,  93)        1.5
# [ 85,  89)        1.75
# [ 81,  85)        2.0
# [ 77,  81)        2.25
# [ 74,  77)        2.5
# [ 0,   74)        3.0

# Run the program repeatedly to test the various different values for input.

def computegrade(score):
  if(score >= 97):
    grade = 1.0
  elif(score >= 93):
    grade = 1.25
  elif(score >= 89):
    grade = 1.5
  elif(score >= 85):
     grade = 1.75
  elif(score >= 81):
    grade = 2.0
  elif(score >= 77):
    grade = 2.25
  elif(score >= 74):
    grade = 2.5
  elif(score >= 0):
    grade = 3.0
  else :
    grade = None

  return grade

score = int(input("Enter score between 0 and 100: "))

display_grade = computegrade(score)
print("Grade: ", display_grade)
```
