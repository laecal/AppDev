What's does math.exp do? What about math.pow?

*Answer:*  

        math.pow is used to calculate a number raised to the power of a number. It takes two arguments: the base and the exponent.
        It specifically deals with floating-point numbers and always returns a floating-point result,
        even if the base and exponent are integers.

        math.exp it returns the natural exponential of a number, meaning e raised to the power of the argument.


How do you calculate logarithms?Logarithms base 10? Natural Logarithms?

*Answer:*

        Logarithms base 10. It represents the exponent to which 10 must be raised to produce a given number x.
        We can calculate common logarithms using the math.log10() function.


```python
# example of logarithmic base 10

import math
result = math.log10(100)
print(result)
```

```python
# example of natural logarithms
import math
result = math.log(100, 2)
print(result)
```
