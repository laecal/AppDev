EXERCISES 1

```python
# Write a program to read through a file and print the contents of the file (line by line) all in upper case.

file_name = input("Enter the name of the file: ")

try:
    with open(file_name, 'r') as file:
        for line in file:
            print(line.upper())
except:
    print(f"File '{file_name}' not found.")
```

EXERCISE 2

```python
# Write a program to read through a file and output the number of lines in a file.

file_name = input("Enter the name of the file: ")

try:
    with open(file_name, 'r') as file:
        for line in file:
            print(line.upper())

    with open(file_name, 'r') as f:
        lines = sum(1 for line in f)
    print("Number of lines: ", lines)

except:
    print(f"File '{file_name}' not found.")
```

```python
# Write a program to prompt for a file name, and then read through the file and look for lines of the form:

# When you encounter a line that starts with "X-DSPAM-Confidence:" pull apart the line to extract
# the floating-point number on the line. Count these lines and then compute the total of the spam
# confidence values from these lines. When you reach the end of the file, print out the average spam confidence.

file_name = input("Enter the file name: ")

total = 0.0
count = 0

try :
    with open(file_name, 'r') as file:
        for line in file:
          if line.startswith("X-DSPAM-Confidence:"):
                confidence = float(line.split(":")[1].strip())
                total += confidence
                count += 1
    if count > 0:
        average = total / count
        print("The average of spam confidence:",average)
    else:
        print("No lines with spam confidence found.")

except:
    print(f"File '{file_name}' not found.")


```
