EXERCISE 1

```python
# Given the code below

# text = 'Business Analytics'

# What will the following expressions produce?

# text[4]
# text[2:4]
# text[:4]
# text[4:]
# text[-2]
# text[-4:-2]
# text[-2:]
# text[:-2]

text = 'Business Analytics'
name = text[4]
name1 = text[2:4]
name2 = text[:4]
name3 = text[4:]
name4 = text[-2]
name5 = text[-4:-2]
name6 = text[-2:]
name7 = text[:-2]

print(name) # this statement will produce 'n' as it is located at the element 4 of the string 'Business Analytics'
print(name1) # this statement will produce 'si'
print(name2) # this statement will produce 'Busi'
print(name3) # this statement will produce 'ness Analytics'
print(name4) # this statement will produce 'c'
print(name5) # this statement will produce 'ti'
print(name6) # this statement will produce 'cs'
print(name7) # this statement will produce 'Business Analyti'
```

EXERCISE 2

```python
# Given the code below:

# phrase = 'One Ring to rule them all'

# What string slicing expression gives out the following words:

# 'One '
# 'Ring'
# 'all'
# 'them'
# 'One Ring to rule them all'

phrase = 'One Ring to rule them all'
p = phrase[0:3] #code for 'One'
h = phrase[4:8] #code for 'Ring'
r = phrase[-3:] #code for 'all'
a = phrase[17:21] #code for 'them'
se = phrase[:] #code for 'One Ring to rule them all'

print(p)
print(h)
print(r)
print(a)
print(se)
```

```python
# Write a program which repeatedly reads a full name, then does the following:
# - strips the name of any trailing spaces
# - capitalizes the first letter
# - puts the rest into smaller letters

while True:
    fullname = input("\nEnter full name: ")
    if fullname.lower() == 'done':
        break
    elif fullname.isnumeric():
        print("invalid input")
    else:
        print(fullname.strip())
        print(fullname.capitalize())
        print(fullname.lower())

```
